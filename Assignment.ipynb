```python
student_marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 89,
    "Eve": 95
}
print("Student Marks:", student_marks)
average_marks = sum(student_marks.values()) / len(student_marks)
print(f"Initial Average Marks: {average_marks:.2f}")
highest_scorer = max(student_marks, key=student_marks.get)
lowest_scorer = min(student_marks, key=student_marks.get)
greater_than_average = {student: mark for student, mark in student_marks.items() if mark > average_marks}
search_name = input("Enter student name to search: ")
if search_name in student_marks:
    print(f"{search_name} scored: {student_marks[search_name]} marks")
    update_name = input("Enter student name to update marks for: ")
    updated_marks = int(input(f"Enter new marks for {update_name}: "))
    student_marks[update_name] = updated_marks
    print(f"Marks for {update_name} updated to {updated_marks}.")
```
